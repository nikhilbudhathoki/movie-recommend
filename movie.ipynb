```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):#converts strings to the python datatypes to real one
        
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):#converts strings to the python datatypes to real one
        
        L.append(i['name'])
        if counter!=3:
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):#converts strings to the python datatypes to real one
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):#function to get root words
    p=[]
    for i in text.split():
        p.append(ps.stem(i))
    return " ".join(p)
        
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer#for the stop words
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
(vectors[5])
```

```python
f=(cv.get_feature_names_out())#gives names of the tokens 5000 which are repeated
```

##### We will not find euclidean distance between the vectors , im highger dim euclidean dis fails
We will use cosine distance which is cosangle between the vectors,More distance isi leass similarity 


```python
from sklearn.metrics.pairwise import cosine_similarity#it is from 0 to 1 if 1 more similarity with less distance
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x : x [1])[1:6]
```

```python
def recommend(movie):
    movie_clean=movie.lower().replace(" ","")
    
    movie_title=pd.DataFrame(new_df['title'].str.replace(" ",""))
        
    movie_index=new_df[movie_title['title'].str.lower()==movie_clean].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x : x [1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('batmanbegins')
```
